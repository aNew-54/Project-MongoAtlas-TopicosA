# Realizando Consultas

```python
from connect import db
from bson.objectid import ObjectId

resultados  = db.libros

# Consultar todo

for libro in resultados.find():
    print(libro)
```

```python
# Consulta que muestre libros mayores al año 1960

for doc in resultados.find({
    'anio': {'$gte':1960}
}):
    print(doc['titulo'], doc['anio'])
```

```python
# Listar los libros de ciencia ficción

for doc in resultados.find({
    'generos':{'$gte': 'ciencia ficción'}
}):
    print(doc['titulo'], doc['anio'])
```

```python


for doc in resultados.find({
    'Stock': {'$gte': 2}
}):
    print(doc['Título'], doc['Anio'], 'Stock:', doc['Stock'], 'Unidades')
```
